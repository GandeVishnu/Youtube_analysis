```python
#plot histoplot for all numeric columns in 'dt'
#count of view_count,likes....
import warnings
warnings.filterwarnings("ignore")
for i in dt.select_dtypes(include="number").columns:
    sns.histplot(data=dt,x=i)
    plt.show()
```

```python
#plot kdeplot for all numeric columns in 'dt'
#count of view_count,likes....
import warnings
warnings.filterwarnings("ignore")
for i in dt.select_dtypes(include="number").columns:
    plt.figure(figsize=(15,6))
    sns.kdeplot(data=dt,x=i)
    plt.show()
```

```python
#plot boxplot for all numeric columns in 'dt'
#count of view_count,likes....
import warnings
warnings.filterwarnings("ignore")
for i in dt.select_dtypes(include="number").columns:
    sns.boxplot(data=dt,x=i)
    plt.show()
```
